```python
import web3 as W3
import urllib.request
import json
```

```python
class Bid(object):
    def __init__(self, BaseTokenAmount = 0, price = 0):
        self.BaseTokenAmount = BaseTokenAmount
        self.price = price

class Ask(object):
    def __init__(self, BaseTokenAmount = 0, price = 0):
        self.BaseTokenAmount = BaseTokenAmount
        self.price = price

def takeBids(market_book_url):
    market = urllib.request.urlopen(market_book_url)
    market = market.read()
    market = market.decode('utf-8')
    market = json.loads(market)
    bids = market['bids']
    collected_bids = []
    for bid in bids:
        bid_to_collect = Bid(float(bid['remainingBaseTokenAmount']), float(bid['price']))
        collected_bids.append(bid_to_collect)
    return collected_bids

def takeAsks(market_book_url):
    market = urllib.request.urlopen(market_book_url)
    market = market.read()
    market = market.decode('utf-8')
    market = json.loads(market)
    asks = market['asks']
    collected_asks = []
    for ask in asks:
        ask_to_collect = Ask(float(ask['remainingBaseTokenAmount']), float(ask['price']))
        collected_asks.append(ask_to_collect)
    return collected_asks
```

```python
markets = urllib.request.urlopen('https://api.radarrelay.com/v2/markets')
markets = markets.read()
markets = markets.decode('utf-8')
markets = json.loads(markets)
data_of_bids = []
data_of_asks = []
#print(markets)
for market in markets:
    #print(market)
    market_id = market['id']
    market_book_url = "https://api.radarrelay.com/v2/markets/" + market_id + "/book"
    data_of_bids.append(takeBids(market_book_url))
    data_of_asks.append(takeAsks(market_book_url))
    #print(market_book_url)
    
    
```

```python
for bids in data_of_bids:
    for market in markets:
        for bid in bids:
            print(market['id'], " ", bid.BaseTokenAmount, " ", bid.price)
```
